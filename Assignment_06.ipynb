Q1. Ask the user to enter a value (e.g., M) for a mathematical function.

 Generate x values from -10 to 10 using np.linspace().

 Compute y values for:
o y= M⋅x2

o y=M⋅sin(x)

 Plot both functions in a single figure:

o Use different colors and line styles.

o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.plot(x, y1, label=r'$y = M \cdot x^2$', color='b', linestyle='-', linewidth=2)
plt.plot(x, y2, label=r'$y = M \cdot \sin(x)$', color='r', linestyle='--', linewidth=2)

plt.legend()
plt.grid(True)
plt.title(f"Graphs of y = M * x^2 and y = M * sin(x) with M = {M}")

plt.show()
```

Q.2 Create a dataset of five subjects and scores.

 Convert it into a Pandas DataFrame.

 Plot the scores using a Seaborn bar plot with:

o Different colors for each bar.

o Annotations on top of each bar.

o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
    'Score': [85, 92, 78, 88, 95]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='Set2')

for index, value in enumerate(df['Score']):
    bar_plot.text(index, value + 1, f'{value}', ha='center', va='bottom', fontsize=12)

plt.title('Scores by Subject', fontsize=16)
plt.xlabel('Subjects', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().

 Create a 2x2 subplot layout displaying:

o Line plot showing cumulative sum.

o Scatter plot with random noise.

 Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 12345
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].plot(np.cumsum(data), label='Cumulative Sum', color='b')
axes[0].set_title('Cumulative Sum of Random Values')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Cumulative Sum')
axes[0].grid(True)

axes[1].scatter(range(50), data, label='Random Noise', color='r')
axes[1].set_title('Scatter Plot with Random Noise')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Random Value')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.

1. Read Total profit of all months and show it using a line plot.

2. Read all product sales data and show it using a multiline plot.

3. Plot bar chart for all the features/attributes.

```python
df_sales = pd.read_csv("https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv")


plt.figure(figsize=(8, 6))
plt.plot(df_sales['month_number'], df_sales['total_profit'], marker='o', linestyle='-', color='b')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid()
plt.show()


plt.figure(figsize=(10, 6))
for column in df_sales.columns[1:-1]:
    plt.plot(df_sales['month_number'], df_sales[column], marker='o', label=column)
plt.title("Product Sales Over Months")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()


plt.figure(figsize=(10, 6))
df_sales.set_index('month_number').plot(kind='bar', figsize=(10, 6))
plt.title("Company Sales Data")
plt.xlabel("Month")
plt.ylabel("Values")
plt.grid()
plt.show()
```
